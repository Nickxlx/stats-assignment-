Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

Answer--> The main difference between a t-test and a z-test lies in the information available about the population standard deviation. 

A t-test is used when the population standard deviation is unknown and is estimated from the sample data. It is typically employed when the sample size is small (typically less than 30).

Example scenario for using a t-test:
Suppose you want to compare the mean heights of two groups of students, one from School A and the other from School B. You collect samples of heights from both groups, and the sample sizes are relatively small (e.g., 25 students from each school). Since the population standard deviation is unknown and the sample sizes are small, you would use a t-test to determine if there is a significant difference between the means of the two groups.

A z-test, on the other hand, is used when the population standard deviation is known or when the sample size is large (typically greater than 30).

Example scenario for using a z-test:
Consider a scenario where you want to compare the mean weights of two groups of individuals, one receiving a new weight loss treatment and the other receiving a placebo. The sample sizes are large, with 100 participants in each group. Additionally, the population standard deviation of weights is known. In this case, you would use a z-test to determine if there is a significant difference in mean weight between the two groups.

Q2: Differentiate between one-tailed and two-tailed tests.

Answer-->
One-Tailed Test:
A one-tailed test, also known as a directional test, is a statistical test where the alternative hypothesis is expressed in a specific direction or one-sided manner. It focuses on whether the sample data falls significantly above or below a certain value, without considering deviations in the opposite direction. The critical region is located on one side of the distribution.

Two-Tailed Test:
A two-tailed test, also known as a non-directional test, is a statistical test where the alternative hypothesis is expressed in a non-specific direction or two-sided manner. It considers deviations in both directions from the null hypothesis value. The critical region is split between the two tails of the distribution.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

Answer--> Type 1 Error:
A Type 1 error, also known as a false positive, occurs when the null hypothesis is rejected when it is actually true.

Example scenario for Type 1 error:
Let's say a medical researcher is testing a new drug's effectiveness for treating a specific disease. The null hypothesis states that the drug has no effect. However, due to random chance or experimental error, the researcher concludes that the drug is effective and rejects the null hypothesis, even though it is actually ineffective. This is a Type 1 error because the researcher falsely concludes that the drug is effective when it is not.

Type 2 Error:
A Type 2 error, also known as a false negative, occurs when the null hypothesis is not rejected when it is actually false.

Example scenario for Type 2 error:
Consider a scenario where a quality control inspector is responsible for determining whether a batch of products meets a certain standard. The null hypothesis is that the batch meets the standard, and the alternative hypothesis is that the batch does not meet the standard. However, due to limitations in the inspection process or a small sample size, the inspector fails to identify that the batch does not meet the standard and fails to reject the null hypothesis. This is a Type 2 error because the inspector fails to detect a true problem with the batch.

Q4: Explain Bayes's theorem with an example.

Answer--> 
Bayes's theorem is a fundamental concept in probability theory that provides a way to update the probability of an event based on new evidence or information.

The formula for Bayes's theorem is:

P(A|B) = ( P(A)) * P(B|A) / P(B)

Where:
P(A|B) is the probability of event A occurring given that event B has occurred.
P(B|A) is the probability of event B occurring given that event A has occurred.
P(A) is the prior probability of event A.
P(B) is the prior probability of event B.

example:

Suppose there is a factory that manufactures electronic components. The factory has two machines, Machine A and Machine B, that produce these components. Machine A is known to have a defect rate of 2%, while Machine B has a defect rate of 5%.

Now, let's say a defective component is randomly selected from the production line, and we want to determine the probability that it was produced by Machine A (P(Machine A|Defective)).

Let's apply Bayes's theorem to calculate this probability:

P(Machine A|Defective) = (P(Machine A) * P(Defective|Machine A) ) / P(Defective)

P(Defective|Machine A) = 0.02 (defect rate of Machine A)
P(Machine A) = 0.5 (assuming equal production from both machines)
P(Defective) = (P(Defective|Machine A) * P(Machine A)) + (P(Defective|Machine B) * P(Machine B))

P(Defective|Machine B) = 0.05 (defect rate of Machine B)
P(Machine B) = 0.5 (assuming equal production from both machines)

P(Defective) = (0.02 * 0.5) + (0.05 * 0.5) = 0.035

Now, we can calculate:

P(Machine A|Defective) = (0.02 * 0.5) / 0.035 ≈ 0.286

This means that the probability of the defective component being produced by Machine A, given that it is defective, is approximately 0.286, or 28.6%.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

Answer--> A confidence interval is a range of values that provides an estimate of the true unknown population parameter, such as the population mean, along with a level of confidence. It indicates the level of uncertainty associated with the estimate.

Example:
Suppose you want to estimate the average height of adult males in a certain population. You collect a random sample of 100 adult males and measure their heights. The sample mean height is 175 cm, and the standard deviation is 5 cm.

You want to calculate a 95% confidence interval for the population mean height.

First, calculate the standard error:
Standard Error = Standard Deviation / sqrt(Sample Size)
Standard Error = 5 / sqrt(100) = 0.5

Next, determine the critical value. Since the sample size is large (100), you can use the Z-distribution and find the critical value associated with a 95% confidence level. This value is approximately 1.96.

Now, calculate the confidence interval:
Confidence Interval = Sample Mean ± (Critical Value * Standard Error)
Confidence Interval = 175 ± (1.96 * 0.5)
Confidence Interval = 175 ± 0.98
Confidence Interval = (174.02, 175.98)

The 95% confidence interval for the average height of adult males in the population is (174.02 cm, 175.98 cm). This means that we can be 95% confident that the true population mean height falls within this range.

Solution:
To solve this problem, we can apply Bayes' Theorem. Let's define the events as follows:

Event A: The component is Type A.
Event D: The component is defective.

We need to find the probability of Event A given Event D, denoted as P(A|D).

According to Bayes' Theorem:

P(A|D) = (P(D|A) * P(A)) / P(D)

Here's how we can calculate each term:

P(D|A) = 0.05 (5% of Type A components are defective)
P(A) = 0.10 (10% of the components produced are Type A)

To calculate P(D), we need to use the law of total probability:

P(D) = P(D|A) * P(A) + P(D|B) * P(B)

P(D|B) = 0.02 (2% of Type B components are defective)
P(B) = 0.90 (90% of the components produced are Type B)

Now, we can substitute the values into Bayes' Theorem:

P(A|D) = (0.05 * 0.10) / ((0.05 * 0.10) + (0.02 * 0.90))
        = 0.005 / (0.005 + 0.018)
        = 0.005 / 0.023
        ≈ 0.217

Therefore, the probability that the defective component is Type A, given that it is defective, is approximately 0.217 or 21.7%.

Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.


Answer--> The margin of error in a confidence interval is the range of values above and below the sample estimate within which the true population parameter is likely to fall. The margin of error is influenced by the confidence level and the variability of the data.

The formula for the margin of error is:

Margin of Error = Critical Value * Standard Error

The critical value corresponds to the desired confidence level and is determined by the distribution. The standard error is calculated based on the sample size and the variability of the data.

In general, as the sample size increases, the margin of error decreases. This is because a larger sample provides more information and reduces the uncertainty associated with the estimate. A larger sample size leads to a more precise estimate of the population parameter and reduces the variability of the data.

Here's an example scenario to illustrate how a larger sample size results in a smaller margin of error:

Scenario:
A company wants to estimate the average customer satisfaction rating for its product. They initially survey a sample of 50 customers and calculate a 95% confidence interval with a margin of error of ±2.5. Later, they decide to increase the sample size to 200 and calculate a new confidence interval.

In the first survey with a sample size of 50, the margin of error is ±2.5. This means that the company is 95% confident that the true average customer satisfaction rating falls within 2.5 points above or below the sample mean.

In the second survey with a sample size of 200, the larger sample provides more information and reduces the uncertainty. As a result, the margin of error decreases. Let's say the new margin of error is ±1.5. This means that with a 95% confidence level, the true average customer satisfaction rating is estimated to be within 1.5 points above or below the sample mean.

By increasing the sample size from 50 to 200, the company has obtained a more precise estimate with a smaller margin of error. The larger sample size allows for a more confident assessment of the population parameter, resulting in a narrower range of values in the confidence interval.

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

Answer--> 

Null hypothesis (H0): The weight loss drug is not significantly effective.
Alternative hypothesis (H1): The weight loss drug is significantly effective.

Significance level: 0.05 (corresponding to a 95% confidence level)

t = (sample mean - hypothesized mean) / (sample standard deviation / √sample size)

Given:

    Sample mean (x̄) = 6 pounds
    Population mean (μ) = 0 (assuming no weight loss)
    Sample standard deviation (s) = 2.5 pounds
    Sample size (n) = 50

Calculating the t-statistic:

    t = (6 - 0) / (2.5 / √50)
    t = 6 / (2.5 / 7.07)
    t = 6 / 1.767
    t ≈ 3.39

Now, we need to compare the calculated t-statistic with the critical value from the t-distribution table. Since we have a large sample size (n > 30), the t-distribution approaches the standard normal distribution.

At a 95% confidence level and degrees of freedom (df) = n - 1 = 50 - 1 = 49, the critical value is approximately ±1.96 (for a two-tailed test).

Since the calculated t-statistic (3.39) is greater than the critical value (1.96), we can reject the null hypothesis.

Conclusion:
Based on the results of the hypothesis test, there is sufficient evidence to conclude that the weight loss drug is significantly effective at a 95% confidence level. 

Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.

Answer--> we can use the formula:

Confidence Interval = Sample Proportion ± Margin of Error

Given: Sample Size (n) = 500 Sample Proportion (p) = 0.65 (65%)

First, we need to calculate the standard error (SE), which is the standard deviation of the sampling distribution of proportions:

    SE = √[(p * (1 - p)) / n]
    SE = √[(0.65 * (1 - 0.65)) / 500]
    SE ≈ √[0.2275 / 500] 
    SE ≈ √0.000455 SE ≈ 0.0213

Next, we calculate the margin of error (ME) using the critical value from the standard normal distribution. For a 95% confidence level, the critical value is approximately 1.96.

    ME = Critical Value * SE
    ME = 1.96 * 0.0213
    ME ≈ 0.0418

Finally, we can calculate the confidence interval:

Confidence Interval = Sample Proportion ± Margin of Error Confidence Interval = 0.65 ± 0.0418 Confidence Interval ≈ (0.6082, 0.6918)

Interpretation: We can be 95% confident that the true proportion of people who are satisfied with their job falls within the range of approximately 60.82% to 69.18%. This means that if we were to repeat the survey multiple times, 95% of the confidence intervals calculated would contain the true proportion of satisfied individuals.

Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.


```python
import scipy.stats as stats

# Sample A statistics
mean_A = 85
stddev_A = 6
sample_size_A = 30

# Sample B statistics
mean_B = 82
stddev_B = 5
sample_size_B = 35

# Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind_from_stats(mean_A, stddev_A, sample_size_A, mean_B, stddev_B, sample_size_B)

# Define the significance level
alpha = 0.01

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("There is a significant difference in student performance between the two teaching methods.")
else:
    print("There is no significant difference in student performance between the two teaching methods.")
```

Confidence Interval = Sample Mean ± (Critical Value * Standard Error)

Given:

    Population Mean (μ) = 60
    Population Standard Deviation (σ) = 8
    Sample Size (n) = 50
    Sample Mean (x̄) = 65
    Confidence Level = 90% (which corresponds to a significance level of α = 0.1)

    SE = σ / √n

we find that the critical value for a 90% confidence level with 49 degrees of freedom is approximately 1.677.

Now we can calculate the confidence interval:

Confidence Interval = Sample Mean ± (Critical Value * Standard Error)
Confidence Interval = 65 ± (1.677 * 1.131)
Confidence Interval ≈ (63.15, 66.85)

Interpretation:
We can be 90% confident that the true population mean falls within the range of approximately 63.15 to 66.85. This means that if we were to repeat the sampling process multiple times and calculate the confidence interval each time, 90% of the intervals would contain the true population mean of 60.

Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

Step 1: State the hypotheses:
Null Hypothesis (H0): Caffeine has no significant effect on reaction time. (μ = μ0)
Alternative Hypothesis (H1): Caffeine has a significant effect on reaction time. (μ ≠ μ0)

Step 2: Set the significance level (α):
In this case, the significance level is given as a 90% confidence level, which corresponds to α = 1 - 0.90 = 0.10.

Step 3: Compute the test statistic:
We will use a t-test since the population standard deviation is unknown. The test statistic for a one-sample t-test is calculated as:

t = (sample mean - hypothesized mean) / (sample standard deviation / sqrt(sample size))

Given:

    Sample Size (n) = 30
    Sample Mean (x̄) = 0.25 seconds
    Sample Standard Deviation (s) = 0.05 seconds
    Hypothesized Mean (μ0) = 0 (no effect)

t = (0.25 - 0) / (0.05 / sqrt(30))
t ≈ 13.416

Step 4: Determine the critical value(s):
Since this is a two-tailed test, we divide the significance level by 2 to get α/2 = 0.10 / 2 = 0.05.

Looking up the critical value in the t-distribution table with (n-1) degrees of freedom (29 degrees of freedom in this case) and α/2 = 0.05, we find that the critical value is approximately ±1.699.

Step 5: Make a decision:
If the test statistic falls outside the critical region (t < -1.699 or t > 1.699), we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.

Since |t| = 13.416 > 1.699, we reject the null hypothesis.

Step 6: Interpret the result:
Based on the sample data, we have evidence to suggest that caffeine has a significant effect on reaction time at a 90% confidence level.
